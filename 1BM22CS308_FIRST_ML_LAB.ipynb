```python
#handling categorical values in the diabets dataset
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_cols)

# Perform One-Hot Encoding
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_data = encoder.fit_transform(df[categorical_cols])

# Create a new DataFrame with encoded columns
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols))

# Concatenate encoded and original DataFrames
df = pd.concat([df, encoded_df], axis=1)

# Drop original categorical columns
df.drop(categorical_cols, axis=1, inplace=True)

# Display the updated DataFrame
df.head()
```

```python
#handling outliers values in dataset diabetes

import pandas as pd

# Load the dataset
df = pd.read_csv("/content/Dataset of Diabetes .csv")

# Calculate quantiles and IQR for 'AGE' column
Q1 = df['AGE'].quantile(0.25)
Q3 = df['AGE'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data to remove outliers
df_no_outliers = df[(df['AGE'] >= lower_bound) & (df['AGE'] <= upper_bound)]
df_no_outliers.head()
# Now, df_no_outliers contains the dataset without outliers in the 'AGE' column.
```

```python
#handling outliers in adult dataset
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/adult.csv")

# Calculate quantiles and IQR for 'AGE' column
Q1 = df['hours-per-week'].quantile(0.25)
Q3 = df['hours-per-week'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter data to remove outliers
df_no_outliers = df[(df['hours-per-week'] >= lower_bound) & (df['hours-per-week'] <= upper_bound)]
df_no_outliers.head()
# Now, df_no_outliers contains the dataset without outliers in the 'AGE' column.
```

```python
import pandas as pd
df=pd.read_csv("/content/Dataset of Diabetes .csv")
from sklearn.preprocessing import MinMaxScaler, StandardScaler
minmax_scaler = MinMaxScaler()
standard_scaler = StandardScaler()
numerical_features = ['AGE', 'Urea', 'Cr', 'HbA1c', 'Chol', 'TG', 'HDL', 'LDL', 'VLDL', 'BMI']
df[numerical_features] = minmax_scaler.fit_transform(df[numerical_features])
df[numerical_features] = standard_scaler.fit_transform(df[numerical_features])
print(df[numerical_features])
```
